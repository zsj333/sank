```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x - 2)**2 * (x + 3)**2
```

```python
def df(x):
    return 2 * (x - 2) * (x + 3)**2 + 2 * (x - 2)**2 * (x + 3)
```

```python
def gradient_descent(f, df, x0, learning_rate=0.01, max_iter=1000, tol=1e-6):
    x = x0
    history = [x]
    
    for i in range(max_iter):
        gradient = df(x)
        x = x - learning_rate * gradient
        history.append(x)
        
        if abs(gradient) < tol:
            break
    
    return x, history
```

```python
x0 = 4  # 初始点
learning_rate = 0.01
max_iter = 1000
tol = 1e-6

min_x, history = gradient_descent(f, df, x0, learning_rate, max_iter, tol)
print(f"全局最小值的近似解: x = {min_x}, f(x) = {f(min_x)}")
```

```python
# 生成x的取值范围
x_vals = np.linspace(-5, 5, 400)
y_vals = f(x_vals)

# 绘制函数图像
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x)')

# 绘制梯度下降路径
history = np.array(history)
plt.plot(history, f(history), 'ro-', label='Gradient Descent Path')

# 标记最小值点
plt.plot(min_x, f(min_x), 'go', label='Global Minimum')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent on a Function with Two Valleys')
plt.legend()
plt.grid(True)
plt.show()
```
